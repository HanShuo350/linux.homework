 我来帮您完整梳理从生成密钥到配置成功的全流程，确保每一步正确衔接：

---

## 完整流程（Windows PowerShell）

### 第1步：生成密钥（如果还没做）

```powershell
# 使用 PowerShell 生成 Ed25519 密钥
ssh-keygen -t ed25519 -C "your_email@example.com"
```

- 提示保存位置：直接回车（默认 `C:\Users\你的用户名\.ssh\id_ed25519`）
- 提示密码：建议设置，或回车留空

---

### 第2步：启动 ssh-agent 服务（关键！）

**以管理员身份运行 PowerShell**，执行：

```powershell
# 设置自动启动并启动服务
Get-Service ssh-agent | Set-Service -StartupType Automatic
Start-Service ssh-agent

# 验证是否运行
Get-Service ssh-agent
```

显示 `Running` 即成功。

---

### 第3步：添加私钥到 ssh-agent

```powershell
ssh-add $env:USERPROFILE\.ssh\id_ed25519
```

成功提示：`Identity added: ...`

---

### 第4步：复制公钥到 GitHub

```powershell
# 显示公钥内容，选中复制
Get-Content $env:USERPROFILE\.ssh\id_ed25519.pub
```

然后：
1. 打开 https://github.com/settings/keys
2. 点击 **New SSH key**
3. 粘贴公钥，保存

---

### 第5步：测试连接

```powershell
ssh -T git@github.com
```

首次连接输入 `yes`，看到 `Hi xxx!` 即成功。
