
## 生成 SHA256 校验文件并验证

```bash
# 1. 创建一个示例文件
echo "这是重要文件内容" > myfile.txt

# 2. 生成 SHA256 校验文件（发布这个）
sha256sum myfile.txt > myfile.txt.sha256

# 3. 验证文件完整性
sha256sum --check myfile.txt.sha256
```

## 完整示例（包含输出）

```bash
# 创建文件
$ echo "Hello World" > test.txt

# 生成哈希
$ sha256sum test.txt
a591a6d40bf420404a011733cfb7b190d62c65bf0bcda32b57b277d9ad9f146e  test.txt

# 保存到校验文件
$ sha256sum test.txt > test.txt.sha256

# 查看校验文件内容
$ cat test.txt.sha256
a591a6d40bf420404a011733cfb7b190d62c65bf0bcda32b57b277d9ad9f146e  test.txt

# 验证文件
$ sha256sum --check test.txt.sha256
test.txt: OK
```
