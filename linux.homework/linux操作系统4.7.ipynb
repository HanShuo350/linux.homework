### 第一步：从 PowerShell 远程连接 Linux

Windows 10/11 自带 SSH 客户端，直接在 PowerShell 里输入以下命令即可连接：

```powershell
ssh 用户名@服务器IP地址
```

*   **示例**：`ssh root@192.168.1.100`
*   **首次连接**：会提示验证主机指纹，输入 `yes` 后回车。
*   **输入密码**：登录密码。**注意**：输入时屏幕不会显示任何字符，这是正常的安全机制，输完按回车键就行。

登录成功后，命令提示符会变成类似 `[root@your-server ~]#` 的形式，说明已成功进入 Linux 系统。

### 第二步：在 Linux 上安装 Node.js

根据你的 Linux 发行版（如 Ubuntu/Debian 或 CentOS），选择对应命令。

*   **Ubuntu/Debian 系统 (使用 apt)**：这是推荐方法，能安装较新的版本。
    ```bash
    # 1. 更新软件包列表
    sudo apt update
    # 2. 安装 Node.js 和 npm（包管理器）
    sudo apt install nodejs npm -y
    ```

*   **CentOS/RHEL 系统 (使用 yum)**：如果默认源版本太老，建议用 **NodeSource** 源安装更新版本。
    ```bash
    # 1. 安装 Node.js 和 npm
    sudo yum install nodejs npm -y
    ```

### 第三步：启动 Node.js REPL 环境

1.  **验证安装**：输入以下命令，如果正确显示版本号，说明安装成功。
    ```bash
    node -v
    npm -v
    ```
2.  **启动 REPL**：在终端里直接输入 `node` 并回车，就会进入 REPL 模式。
    ```bash
    node
    ```
3.  **测试代码**：进入 `>` 提示符后，就可以输入 JavaScript 代码了。例如：
    ```javascript
    > console.log("Hello from Linux!");
    Hello from Linux!
    undefined
    >
    ```
4.  **退出 REPL**：按两次 `Ctrl+C`，或者输入 `.exit` 后回车。
